Regular expressions
attrs:Converts attributes into dictionary form

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

html=urlopen('http://www.pythonscraping.com/pages/page3.html')
bs=BeautifulSoup(html, 'html.parser')

for a in bs.find_all('img', {'src':re.compile('\.\.\/img\/gifts\/img(.)*\.jpg')}):
  #print (a.attrs['src'])
  print(a.attrs, len(a.attrs))
```

Lambda function


```python
a=lambda b:b**2
s=lambda b:len(b)
print(a(2))
print(s('ttqqweqwet'))
```

Lambda expression

Input is a tag and output is boolean
So BeautifulSoup returns the tags that satisfy the condition

```python
l=bs.find_all(lambda tag: len(tag.attrs)==2)
print('Number of tags where number of attributes is 2:',len(l))

for a in l:
  print(a,'\n\n')
```

Accessing Websites that block Web Crawling

```python
from urllib.request import Request
#html=urlopen('https://www.worldometers.info/coronavirus/')
html=urlopen(Request('https://www.worldometers.info/coronavirus/',headers={'User-Agent':'Mozilla'}))
bs=BeautifulSoup(html, 'html.parser')
print(bs)
```

Dealing with Children and Descendants



```python
html=urlopen('http://www.pythonscraping.com/pages/page3.html')
bs=BeautifulSoup(html,'html.parser')
#print(bs)
for child in bs.find('table',{'id':'giftList'}).children:
  print(child)
```
